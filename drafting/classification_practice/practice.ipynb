```python
import sklearn as sk
import pandas as pd
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d

from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
```

```python
df = pd.read_csv('Iris.csv')
df
```

```python
df['species num'] = df['Species'].map({'Iris-setosa':0, 'Iris-versicolor':1.0, 'Iris-virginica':2.0})
df
```

```python
X = df[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]]
Y = df["Species"]
```

```python
# train - test split using sklearn
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

print(f'{len(X_train)} training size, {len(X_test)} testing size')
```

```python
# Scale the data using a standard scaler - converts to z-score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score

scaler = StandardScaler()
scaler.fit(X_train)
X_train_zscores = scaler.transform(X_train)
X_test_zscores = scaler.transform(X_test)

# Note that since the labels for classification are strings, they should not be transformed
```

```python
# Apply Support Vector Classification 
from sklearn.svm import SVC

svm_classifier = SVC(kernel='rbf', random_state=42, gamma='scale')
svm_classifier.fit(X_train_zscores, Y_train)

training_scores = svm_classifier.score(X_train_zscores, Y_train)
print(f'Accuracy on training data: {training_scores}')

testing_scores = svm_classifier.score(X_test_zscores, Y_test)
print(f'Accuracy on testing data: {testing_scores}')
```

```python
# Apply K-Nearest Neighbors Classifier
from sklearn.neighbors import KNeighborsClassifier

knn_classifier = KNeighborsClassifier(weights='distance')
knn_classifier.fit(X_train_zscores, Y_train)

training_scores = knn_classifier.score(X_train_zscores, Y_train)
print(f'Accuracy on training data: {training_scores}')


testing_scores = knn_classifier.score(X_test_zscores, Y_test)
print(f'Accuracy on testing data: {testing_scores}')
```

```python
# Use XGBoost as Classifier
import xgboost as xgb

xgb_classifier = xgb.XGBClassifier()
xgb_classifier.fit(X_train_zscores, Y_train)

training_scores = xgb_classifier.score(X_train_zscores, Y_train)
print(f'Accuracy on training data: {training_scores}')


testing_scores = xgb_classifier.score(X_test_zscores, Y_test)
print(f'Accuracy on testing data: {testing_scores}')
```
